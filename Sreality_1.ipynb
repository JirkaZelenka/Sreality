* Nedostatek - nemám souřadnice - > nemám pořádné adresy -> nemám dobrá města, okresy, kraje

```python

# scrappig elements, které v sobě mají Lokaci, Velikost, Cenu

def get_soup_elements(typ_obchodu = "prodej", typ_stavby = "byty", pages = 5):  
    
    browser = webdriver.Chrome()
    
    ##########################################
    # 1. Volba Prodej/Pronájem, Byty/Domy,                 --Aukce/Bez Aukce (jen pro Prodeje) zatím nechávám být, cpe se mi to doprostřed url
    ##########################################    
    
    url_x = r"https://www.sreality.cz/hledani"             # Všechny byty, lepší než naklikat všechno, kratší url a přehlednější
    url = url_x + "/" +  typ_obchodu + "/" +  typ_stavby
    
    ##########################################
    # 2. načtení webu
    ##########################################
    browser.get(url) # (url).text ??
    sleep(random.uniform(1.0, 1.5))
    innerHTML = browser.execute_script("return document.body.innerHTML")
    soup = BeautifulSoup(innerHTML,'lxml') # "parser" ??
    elements = soup.find_all(class_ = "text-wrap")     # !!!!!!!!!!!!!!!!! změněno, protože H2 neobsahovalo všechny věci, jen nadpisek.
    
    ##########################################
    # 3. zjištění počtu listů - mělo by být optional, ale nevadí
    ##########################################
    records = soup.find_all(class_ ='numero ng-binding')[1].text
    records = re.split(r'\D', str(records))                         # Tady byla chyba, kdoví proč "pages" místo "records", což mi to podělávalo
    records = ",".join(records).replace(",", "")
    records = int(records)
    max_page = math.ceil(records / 20)   # 12 072 inzerátů  = 604 stránek, správně
    print("----------------")
    print("Scrapping: " + str(typ_obchodu) + " " + str(typ_stavby))
    print("Total records: " + str(records))
    print("Total pages: " + str(max_page))
    
    ##########################################
    # 4. nastavení počtu stránek  -mělo být víc promakané
    ##########################################
    if pages == 999:      # (NE)Speciální případ, chci všechny inzeráty - standardně tu bude asi kolem 600 stránek max, takže 999 je volné k použití
        pages = max_page
    
    print("I scrap (only) " + str(pages) + " pages.")
    print("----------------")
    
    ##########################################
    # 4. Scrapping zbylých listů - naštěstí v jednom okně
    ##########################################    
    
    for i in range(pages-1):   # pages = 650 -> range-1 = (0-648) -> i+2 = (2-650) ->  pages = 2-650
        i = i+2
        
        print("Page " + str(i-1) + " = " + str(round(100*(i-1)/(pages), 2)) + "% progress.")
        print("Estimated time remaining: " + str(round(random.uniform(3.3, 3.85)*(pages-(i-1)), 2 )) + " seconds.")  # čas kolem 3.5, spíš maličko nahoru. Odhadnuto podle Scrappingu 600 stránek
        
        url2 = url + "?strana=" + str(i)
        browser.get(url2)

        sleep(random.uniform(1.0, 1.5))

        innerHTML = browser.execute_script("return document.body.innerHTML")
        soup2 = BeautifulSoup(innerHTML,'lxml') # "parser" ??
        elements2 = soup2.find_all(class_ = "text-wrap")    # !!!!!!!!!!!!!!!!! změněno, protože H2 neobsahovalo všechny věci, jen nadpisek.
        
        elements = elements + elements2
        # tyto se už můžou posčítat, naštěstí, řpedtím než z nich budeme dělat elems = prvky třeba jména
        
    browser.quit()   # novinka :)
    
    return elements

######################################################
#######################################################

"""
x = 999
elements_prodej_byty = get_soup_elements(typ_obchodu = "prodej", typ_stavby = "byty", pages = x)
elements_pronajem_byty = get_soup_elements(typ_obchodu = "pronajem", typ_stavby = "byty", pages = x)
elements_prodej_domy = get_soup_elements(typ_obchodu = "prodej", typ_stavby = "domy", pages = x)
elements_pronajem_domy = get_soup_elements(typ_obchodu = "pronajem", typ_stavby = "domy", pages = x)
"""
#
```

## 4a) Pro Scrapping 1 - čištění dat

```python

def cleaning(data):


    ##########################################
    # 1. Split názvu na jednotlivé parametry
    ##########################################
    data[['buy_rent','home_house', "rooms", "meters", "m","0", "1", "2", "3"]] =  data.name.str.split(expand=True)  # Vytvoří se 4 idiotské sloupce navíc, nevím proč
    data = data[["name", "locality", "price", 'buy_rent','home_house', "rooms", "meters"]]

    ##########################################
    # 2. Dropnutí nulových cen
    ##########################################
    display(len(data))
    data["price"] = data["price"].fillna(0)    #
    data = data.loc[~(data.price ==0)]
    display(len(data)) 

    
    data["price"] = (data["price"]/1000).astype(int)    
    # Převedení millionů na tisíce - snad přehlednější
    
    ##########################################
    # 3. 6+ Pokojů doplnění metrů z názvu (odhaleno díky Unique hodnotám - 40x "pokojů")
    ##########################################
    def six(x):
        x = re.findall(r'\d{3}', str(x))
        x = ''.join(x)
        return x

    data.loc[data["rooms"] == "6", "meters"] = data.loc[data["rooms"] == "6", "name"].map(six)
    # Komplikované, ale praktické
    
    ##########################################
    # 4. Města, Okresy
    ##########################################
  
    # Potřebuji detekovat čárku s mezerou a pokračovat buĎ k mezeře s pomlčkou nebo ke konci stringu... (víceslovná města)
    def city(x):
        x = re.split(r'\W\s', x)   # (W + s) == non-letter(-) + space // případně by to chtělo (s + W) == space + (-) a definovatten znak, aby to nebralo tečky !!!!!!
        try:
            x = x[1]    # TOTO jsem chtěl ,ale některé adresy mají jen jednu část - Město (snad)
        except:
            x = x[0]
        return x
    
    data["city"] = data["locality"].map(city)
    
    # Pomocný sloupec kde beru jen první slovo z názvu "city", což mi zajistí slovo Praha místo Praha_X
    def praha(x):
        x = re.findall(r'\w+', x)[0]
        return x
    
    data["city_one_praha"] = data["city"].map(praha)
    # A tady dokopíruju ostatní názvy jiné než Praha, a mám hotovyý sloupec
    data.loc[data["city_one_praha"] != "Praha", "city_one_praha"] = data["city"]

    
    
    return data

######################################################
#######################################################
data = cleaning(data)
data
# 12.440 -> 11.599
```
